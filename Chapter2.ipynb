# 2.2 MovieLens의 영화 평점 데이터

MovieLes 1M(백만 개) 데이터셋  
약 6,000명의 사용자들로부터 4,0000여 편의 영화에 대한 백만 개의 영화 평점을 담고 있다.  
평점, 사용자 정보, 영화 정보, 이 세가지 테이블로 구성되어 있다.

```python
import pandas as pd
import os
encoding = 'latin1'

upath = os.path.expanduser('./data/movielens/users.dat')
rpath = os.path.expanduser('./data/movielens/ratings.dat')
mpath = os.path.expanduser('./data/movielens/movies.dat')

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
mnames = ['movie_id', 'title', 'genres']

users = pd.read_csv(upath, sep='::', header=None, names=unames, encoding=encoding)
ratings = pd.read_csv(rpath, sep='::', header=None, names=rnames, encoding=encoding)
movies = pd.read_csv(mpath, sep='::', header=None, names=mnames, encoding=encoding)
```

나이와 성별에 따른 어떤 영화의 평균평점을 계산해보자.

```python
# merge함수를 사용하여 ratings테이블과 users테이블을 병합하고, 그 결과를 movies테이블과 병합
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data
```

```python
data.ix[0]
```

```python
#성별에 따른 각 영화의 평균 평점을 구하려면 pivot_table 매서드를 사용
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
```

```python
mean_ratings[:5]
```

```python
# 데이터를 영화제목으로 그룹화하고 xize()함수를 사용해서 제목별 평점 정보 건수를 Series객체로 얻어냄
ratings_by_title = data.groupby('title').size()
```

```python
ratings_by_title[:10]
```

```python
# 250건 이상의 평점 정보가 있는 영화만 추림
active_titles = ratings_by_title.index[ratings_by_title >= 250]
```

```python
active_titles
```

```python
# 250건 이상의 평점 정보가 있는 영화에 대한 색인은 mean_ratings에서 항목을 선택하기 위해 사용
mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings
```

```python
# 여성에게 높은 평점을 받은 영화 목록을 확인하기 위해 F열을 내림차순으로 정렬
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)
```

```python
top_female_ratings
```

## 2.2.1 평점 차이 구하기

```python
# 평균 평점 차이를 나타내는 열을 추가
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
# 평균 평점 차이를 기준으로 정렬(여성들이 더 선호하는 순위)
sorted_by_diff = mean_ratings.sort_values(by='diff')
```

```python
sorted_by_diff[:15]
```

```python
# 정렬된 행에서 역순으로 15개 행을 잘라냄(남자가 더 선호하는 순위)
sorted_by_diff[::-1][:15]
```

성별에 관계없이 영화에 대한 호불호가 극명하게 나뉘는 영화를 찾아보자.  
호불호는 평점의 분산이나 표준편차를 통해 측정

```python
# 평점의 표준편차
rating_std_by_title = data.groupby('title')['rating'].std()
```

```python
# active_titles만 선택
rating_std_by_title = rating_std_by_title.ix[active_titles]
```

```python
# 내림차순으로 정렬
rating_std_by_title.sort_values(ascending=False)[:10]
```
